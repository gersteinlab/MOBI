```python
import os
```

```python
samples = [ii.replace(".tar.gz", "") for ii in os.listdir("./") if ii.endswith(".tar.gz")]
```

```python
for ss in samples:
    with open("./%s.md" % ss, "w") as ff:
        motifs = [ii.replace(".meme", "") for ii in os.listdir("./%s/" % ss)]
        ff.write("[Download all in one tar.gz](https://github.com/gersteinlab/MOBI/raw/master/download/%s.tar.gz)  \n" % ss)
        for mm in motifs:
            ff.write("[{mm}](https://raw.githubusercontent.com/gersteinlab/MOBI/master/download/{ss}/{mm}.meme)\n".format(mm=mm, ss=ss))
```
