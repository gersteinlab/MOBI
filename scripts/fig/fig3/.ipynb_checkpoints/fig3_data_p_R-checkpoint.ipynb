```r
nbp_list = c(10, 20, 40, 60, 80, 100, 120, 140, 160, 180, 200)
rank_list = c('RankSPP',
 'RankLinear_0.1',
 'RankLinear_0.2',
 'RankLinear_0.3',
 'RankLinear_0.4',
 'RankLinear_0.5',
 'RankLinear_0.6',
 'RankLinear_0.7',
 'RankLinear_0.8',
 'RankLinear_0.9',
 'RankLinear_1.0',
 'RankLinear_2.0',
 'RankLinear_3.0',
 'RankLinear_4.0',
 'RankLinear_5.0',
 'RankLinear_6.0',
 'RankLinear_7.0',
 'RankLinear_8.0',
 'RankLinear_9.0',
 'RankLinear_10.0',
 'RankCrowdness')

tool = "DREME"
top = "top5"
```

```r
get_p <- function(tomtom_summary_dir, tool, top, nbp_list, rank_list) {
    
    filename = paste(tomtom_summary_dir, "/", tool, "_RankSPP_100_", top, ".txt", sep="")
    spp = read.csv(filename, sep="\t")[["accuracy"]]
    
    nn_data = list()
    for(nn in nbp_list) {
        rank_data = list()
        for(rank in rank_list) {
            filename = paste(tomtom_summary_dir, "/", tool, "_", rank, "_", nn, "_", top, ".txt", sep="")
            data = read.csv(filename, sep="\t")[["accuracy"]]
            wilcox.result = wilcox.test(data, spp, paired = T, alternative = "greater")
            rank_data = c(rank_data, wilcox.result$p.value)
        }
        nn_data = c(nn_data, rank_data)
    }
    df <- data.frame(matrix(unlist(nn_data), nrow=11, byrow=T))
    colnames(df) = rank_list
    rownames(df) = nbp_list
    df_star = data.frame(df)
    df_star[df<0.05] = "*"
    df_star[df>=0.05] = ""
    return(df_star)
}
```

```r
options(warn=-1)
df1 = get_p("~/slayman/motif_inference/result/UnifyParaSearch/fly/tomtom_summary/", tool, top, nbp_list, rank_list)
df2 = get_p("~/slayman/motif_inference/result/UnifyParaSearch/worm/tomtom_summary/", tool, top, nbp_list, rank_list)
df3 = get_p("~/slayman/motif_inference/result/UnifyParaSearch/humanGM12878/tomtom_summary/", tool, top, nbp_list, rank_list)
df4 = get_p("~/slayman/motif_inference/result/UnifyParaSearch/humanK562/tomtom_summary/", tool, top, nbp_list, rank_list)
options(warn=0)
```

```r
write.table(df1, file="fig3_data_fly_p.txt", sep="\t", quote=F)
write.table(df2, file="fig3_data_worm_p.txt", sep="\t", quote=F)
write.table(df3, file="fig3_data_GM12878_p.txt", sep="\t", quote=F)
write.table(df4, file="fig3_data_K562_p.txt", sep="\t", quote=F)
```
