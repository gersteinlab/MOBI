```python
color1 = (150/255,84/255,84/255,0.9)
color2 = (199/255,184/255,161/255,0.9)
color3 = (156/255,168/255,184/255,0.9)

color4 = (8/255,186/255,220/255,0.7)
color5 = (245/255,245/255,245/255,1)
color6 = (200/255,153/255,136/255,0.7)
```

```python
def get_plt_data(data_type):

    df1 = pd.read_csv("/home/jg2447/slayman/motif_inference/result/UnifyParaSearch/%s/performance/DREME_top5.txt" % data_type, sep="\t", header=None)
    stats1 = df1.mean().values

    df2 = pd.read_csv("/home/jg2447/slayman/motif_inference/result/UnifyParaSearch/%s/performance/HOMER_top5.txt" % data_type, sep="\t", header=None)
    stats2 = df2.mean().values

    df3 = pd.read_csv("/home/jg2447/slayman/motif_inference/result/UnifyParaSearch/%s/performance/MEME_top5.txt" % data_type, sep="\t", header=None)
    stats3 = df3.mean().values
    return(df1, df2, df3, stats1, stats2, stats3)
```

```python
figsize = (3.2, 3.2)
panel_number_fs = 8
x_tick_label_fs = 6
y_tick_label_fs = 6
x_label_fs = 7
y_label_fs = 7
title_fs = 6
legend_fs = 5
```

```python
sns.set_context("paper")

fig, axs = plt.subplots(
    nrows=2, ncols=2, 
    sharex=False, sharey=False,
    gridspec_kw={'hspace': 0, 'wspace': 0, "height_ratios":[1, 1]},
    figsize=figsize, dpi=300)

names=['Increase', 'No change', 'Decrease']

#### A
ax = axs[0,0]

df1, df2, df3, stats1, stats2, stats3 = get_plt_data("fly")
ax.pie(
    stats1[2:5], 
    labels=None,
    colors=[color4, color5, color6],
    textprops={'fontsize': 14},
    radius=1,
    counterclock=False,
    wedgeprops=dict(width=0.25, edgecolor='w'),
    startangle=90)
ax.pie(
    stats2[2:5],
    labels=None,
    colors=[color4, color5, color6],
    textprops={'fontsize': 14},
    radius=0.75,
    counterclock=False,
    wedgeprops=dict(width=0.25, edgecolor='w'),
    startangle=90)
ax.pie(
    stats3[2:5],
    labels=None,
    colors=[color4, color5, color6],
    textprops={'fontsize': 14},
    radius=0.5,
    counterclock=False,
    wedgeprops=dict(width=0.25, edgecolor='w'),
    startangle=90)

ax.annotate("DREME", xy=(0.025,0.83), ha='left', va='center', rotation=-10, fontsize=4)
ax.annotate("HOMER", xy=(0.025,0.57), ha='left', va='center', rotation=-10, fontsize=4)
ax.annotate("MEME", xy=(0.025,0.32), ha='left', va='center', rotation=-10, fontsize=4)

ax.set_title("Fly", fontsize=title_fs, y=0.9)

#### B
ax = axs[0,1]

df1, df2, df3, stats1, stats2, stats3 = get_plt_data("worm")
ax.pie(
    stats1[2:5], 
    labels=None,
    colors=[color4, color5, color6],
    textprops={'fontsize': 14},
    radius=1,
    counterclock=False,
    wedgeprops=dict(width=0.25, edgecolor='w'),
    startangle=90)
ax.pie(
    stats2[2:5],
    labels=None,
    colors=[color4, color5, color6],
    textprops={'fontsize': 14},
    radius=0.75,
    counterclock=False,
    wedgeprops=dict(width=0.25, edgecolor='w'),
    startangle=90)
ax.pie(
    stats3[2:5],
    labels=None,
    colors=[color4, color5, color6],
    textprops={'fontsize': 14},
    radius=0.5,
    counterclock=False,
    wedgeprops=dict(width=0.25, edgecolor='w'),
    startangle=90)

ax.annotate("DREME", xy=(0.025,0.83), ha='left', va='center', rotation=-10, fontsize=4)
ax.annotate("HOMER", xy=(0.025,0.57), ha='left', va='center', rotation=-10, fontsize=4)
ax.annotate("MEME", xy=(0.025,0.32), ha='left', va='center', rotation=-10, fontsize=4)

ax.set_title("Worm", fontsize=title_fs, y=0.9)

#### C
ax = axs[1,0]

df1, df2, df3, stats1, stats2, stats3 = get_plt_data("humanGM12878")
ax.pie(
    stats1[2:5], 
    labels=None,
    colors=[color4, color5, color6],
    textprops={'fontsize': 14},
    radius=1,
    counterclock=False,
    wedgeprops=dict(width=0.25, edgecolor='w'),
    startangle=90)
ax.pie(
    stats2[2:5],
    labels=None,
    colors=[color4, color5, color6],
    textprops={'fontsize': 14},
    radius=0.75,
    counterclock=False,
    wedgeprops=dict(width=0.25, edgecolor='w'),
    startangle=90)
ax.pie(
    stats3[2:5],
    labels=None,
    colors=[color4, color5, color6],
    textprops={'fontsize': 14},
    radius=0.5,
    counterclock=False,
    wedgeprops=dict(width=0.25, edgecolor='w'),
    startangle=90)

ax.annotate("DREME", xy=(0.025,0.83), ha='left', va='center', rotation=-10, fontsize=4)
ax.annotate("HOMER", xy=(0.025,0.57), ha='left', va='center', rotation=-10, fontsize=4)
ax.annotate("MEME", xy=(0.025,0.32), ha='left', va='center', rotation=-10, fontsize=4)

ax.set_title("GM12878", fontsize=title_fs, y=0.9)

#### D
ax = axs[1,1]

df1, df2, df3, stats1, stats2, stats3 = get_plt_data("humanK562")
ax.pie(
    stats1[2:5], 
    labels=None,
    colors=[color4, color5, color6],
    textprops={'fontsize': 14},
    radius=1,
    counterclock=False,
    wedgeprops=dict(width=0.25, edgecolor='w'),
    startangle=90)
ax.pie(
    stats2[2:5],
    labels=None,
    colors=[color4, color5, color6],
    textprops={'fontsize': 14},
    radius=0.75,
    counterclock=False,
    wedgeprops=dict(width=0.25, edgecolor='w'),
    startangle=90)
ax.pie(
    stats3[2:5],
    labels=None,
    colors=[color4, color5, color6],
    textprops={'fontsize': 14},
    radius=0.5,
    counterclock=False,
    wedgeprops=dict(width=0.25, edgecolor='w'),
    startangle=90)

ax.annotate("DREME", xy=(0.025,0.83), ha='left', va='center', rotation=-10, fontsize=4)
ax.annotate("HOMER", xy=(0.025,0.57), ha='left', va='center', rotation=-10, fontsize=4)
ax.annotate("MEME", xy=(0.025,0.32), ha='left', va='center', rotation=-10, fontsize=4)

ax.set_title("K562", fontsize=title_fs, y=0.9)
ax.legend(names, fontsize=5, loc='center left', bbox_to_anchor=(-0.25, 1.0), frameon=False)

plt.savefig("./figS1.pdf", dpi="figure", bbox_inches="tight")
```
