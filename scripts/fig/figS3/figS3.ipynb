```python
import os
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
import seaborn as sns
import sys
sys.path.append("/home/jg2447/slayman/motif_inference/MOBI/scripts/")
from src import plt_func
from src import stats
```

```python
enrichment_DNAinput_fly = pd.read_csv("/home/jg2447/slayman/motif_inference/result/UnifyParaSearch/fly/enrichment/RankSPP_100.txt", sep="\t", index_col=0).loc[:,["250"]]
enrichment_mockIP_fly = pd.read_csv("/home/jg2447/slayman/motif_inference/result/mockIP/fly/enrichment/RankSPP_100.txt", sep="\t", index_col=0).loc[:,["250"]]
fly = pd.merge(enrichment_DNAinput_fly, enrichment_mockIP_fly, left_index=True, right_index=True)
enrichment_DNAinput_fly = fly.iloc[:, 0].values
enrichment_mockIP_fly = fly.iloc[:, 1].values
```

```python
enrichment_DNAinput_worm = pd.read_csv("/home/jg2447/slayman/motif_inference/result/UnifyParaSearch/worm/enrichment/RankSPP_100.txt", sep="\t", index_col=0).loc[:,["250"]]
enrichment_mockIP_worm = pd.read_csv("/home/jg2447/slayman/motif_inference/result/mockIP/worm/enrichment/RankSPP_100.txt", sep="\t", index_col=0).loc[:,["250"]]
worm = pd.merge(enrichment_DNAinput_worm, enrichment_mockIP_worm, left_index=True, right_index=True)
enrichment_DNAinput_worm = worm.iloc[:, 0].values
enrichment_mockIP_worm = worm.iloc[:, 1].values
```

```python
from scipy.stats import wilcoxon
```

```python
wilcoxon(enrichment_mockIP_fly, enrichment_DNAinput_fly, alternative='greater')
```

```python
wilcoxon(enrichment_mockIP_worm, enrichment_DNAinput_worm, alternative='greater')
```

```python
figsize = (1, 1.25)
panel_number_fs = 8
x_tick_label_fs = 5
y_tick_label_fs = 5
x_label_fs = 5
y_label_fs = 5
title_fs = 6
legend_fs = 5

colors = ["#8ff098", "#8ba898"]

sns.set_context("paper")

fig = plt.figure(figsize=figsize, dpi=300)
ax = plt.gca()

ax.bar(
    [0,1,2.5,3.5],
    [enrichment_DNAinput_fly.mean(), enrichment_mockIP_fly.mean(), enrichment_DNAinput_worm.mean(), enrichment_mockIP_worm.mean()],
    color = ["#ed6a6d", "#ffffff", "#ed6a6d", "#ffffff"], edgecolor="#ed6a6d")

#ax.set_ylim([-0.75,3.75])
ax.yaxis.set_ticks([0, 0.1, 0.2])
ax.get_yaxis().set_tick_params(size=2, width=0.6, pad=0.5, labelsize=x_tick_label_fs)   
ax.set_ylabel("Motif enrichment", fontsize=y_label_fs, labelpad=0.3)
ax.xaxis.set_ticks([0.5, 3])
ax.get_xaxis().set_ticklabels(["Fly", "Worm"])
ax.get_xaxis().set_tick_params(size=0, pad=2, labelsize=x_tick_label_fs)   


ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines["left"].set_linewidth(0.5)
ax.spines["bottom"].set_visible(False)

ax.legend([ax.patches[0], ax.patches[1]], ["DNA input", "Mock IP"], fontsize=legend_fs-1, frameon=False, bbox_to_anchor=(0.45,1.02), handlelength=1.5)

plt_func.ax_trim(ax)

plt.savefig("./figS3.pdf", dpi="figure", bbox_inches="tight")
```
